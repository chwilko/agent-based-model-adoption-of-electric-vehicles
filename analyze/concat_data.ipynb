```python
import os
import pandas as pd
import numpy as np
```

```python
euro_file = os.sep.join(["..", "data", "eurpln_d.csv"])
get_full_path = lambda x: os.sep.join(["..", "data", "energy_row", x])
```

```python
datas = sorted(os.listdir(os.sep.join(["..", "data", "energy_row"])))
```

```python
df_euro = pd.read_csv(euro_file)
```

```python
column = ["full_data", "DA", "currency", "empty"]


def set_cols(df):
    df.columns = column
    return df


df = pd.concat(
    [set_cols(pd.read_csv(get_full_path(path))) for path in datas], ignore_index=True
)

# def set_data(x):
#     x =


#     pass


df["data"] = df["full_data"].apply(lambda x: "-".join(x[:10].split(".")[::-1]))
# df["data"] = pd.to_datetime(df["data"])
df = df[df["currency"].notna() & df["DA"].notna()]


def to_number(x):
    try:
        return float(x)
    except:
        None


df["DA"] = df["DA"].apply(lambda x: to_number(x))


# def set_PLN(row: pd.Series):
#     if row.currency == "PLN":
#         return row.DA
#     try:
#         today = str(row.data)
#         # print(today)
#         return row.DA * df_euro.Zamkniecie[df_euro.Data == today]
#     except IndexError:
#         print("Ass")
#         return None

# df["DA"] = df.apply(
#     set_PLN,
#     axis=1,
# )

df["DA"] = df.apply(
    lambda row: row["DA"]
    if row["currency"] == "PLN"
    else df_euro.loc[df_euro["Data"] == row["data"], "Zamkniecie"].values[0]
    if df_euro["Data"].isin([row["data"]]).any()
    else None,
    axis=1,
)

df["currency"] = df.apply(
    lambda row: "PLN",
    axis=1,
)
```
