# Analyze car paremeters

I use data from [otomoto.pl](https://www.otomoto.pl/osobowe) acces on filename (DD_MM_YYYY_hh_mm)

Main goal of analyze is found driver profiles for model.

```python
import os
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utils import add_path, get_path_with

add_path()
from model import CV, EV, PHEV

car_types = [CV, PHEV, EV]
```

### set plot params

```python
# plt.style.use("dark_background")
plt.style.use("grayscale")
# plt.rc("grid", alpha=0.3)
kde_params = {
    "fill": True,
    "linewidth": 1,
    "common_norm": False,
}

hist_params = {
    "fill": True,
    "linewidth": 0,
    "common_norm": False,
    "common_bins": False,
    "kde": True,
    "stat": "density",
}
```

### Set data

```python
data_file = os.sep.join([get_path_with("data"), "data", "cars02_05_2023_01_21.csv"])
df = pd.read_csv(data_file)

aliases = {
    "petrol": CV,
    "diesel": CV,
    "petrol-lpg": CV,
    "petrol-cng": CV,
    "electric": EV,
    "hybrid": PHEV,
}

df["type"] = df["fuel_type"].apply(lambda x: aliases[x])

# today_year = datetime.date.today().year
today_year = 2023
df["age"] = df["year"].apply(lambda x: today_year - x)

df["mean_year_miliage"] = df.apply(
    lambda row: row["mileage"] // row.age if row.age != 0 else None, axis=1
)
df = df[df["age"] < 30]
```

## Mean miliage

```python
suggest_profiles = {
    CV: 14000,
    PHEV: 8000,
    EV: 3000,
}
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(df, y="age", x="type")

plt.title("Car age");
```

```python
max_age = 5
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
max_age_median = max_age

median_mean_miliage = {
    CV: df[(df["type"] == CV) & (df["age"] <= max_age_median)][
        "mean_year_miliage"
    ].median(),
    EV: df[(df["type"] == EV) & (df["age"] <= max_age_median)][
        "mean_year_miliage"
    ].median(),
    PHEV: df[(df["type"] == PHEV) & (df["age"] <= max_age_median)][
        "mean_year_miliage"
    ].median(),
}

hist_plot = sns.histplot(
    df[(df["mean_year_miliage"] < 75_000) & (df["age"] <= max_age_median)],
    x="mean_year_miliage",
    hue="type",
    **hist_params,
)


for key in suggest_profiles:
    plt.axvline(x=median_mean_miliage[key], color="m")


plt.xlim([None, 35_000])
plt.title("histplot of mean year miliage")
plt.xlabel("mean year miliage");
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
sns.violinplot(
    df[(df["mean_year_miliage"] < 100000) & (df["age"] < max_age)],
    y="mean_year_miliage",
    x="type",
    order=[CV, PHEV, EV],
)


for i, key in enumerate(car_types):
    plt.axhline(y=median_mean_miliage[key], xmin=i / 3, xmax=(i + 1) / 3)

plt.ylabel("average annual miliage")
plt.title("average annual miliage")
```

# Profiles distribution

```python
sns.violinplot(df[(df["age"] < max_age)], y="mean_year_miliage")
for key in suggest_profiles:
    plt.axhline(y=median_mean_miliage[key], color="m")
```

```python
profiles_distribution = {}
k = 9

profiles_distribution[CV] = len(
    df[
        (df["age"] < max_age)
        & (
            df.mean_year_miliage
            > (median_mean_miliage[CV] * k + (10 - k) * median_mean_miliage[PHEV]) / 10
        )
    ]
) / len(df[(df["age"] < max_age)])

profiles_distribution[EV] = len(
    df[
        (df["age"] < max_age)
        & (
            df.mean_year_miliage
            < (median_mean_miliage[EV] * k + (10 - k) * median_mean_miliage[PHEV]) / 10
        )
    ]
) / len(df[(df["age"] < max_age)])

profiles_distribution[PHEV] = 1 - profiles_distribution[CV] - profiles_distribution[EV]
```
